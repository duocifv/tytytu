```python
# @title Di chuyển các file Wan 2.1 đã tải sang Google Drive
import os

# Đường dẫn nguồn (Thư mục tạm của Colab)
SOURCE_BASE_DIR = "/content/drive/MyDrive/ComfyUI/models"
# Đường dẫn đích (Thư mục ComfyUI trên Drive)
DEST_BASE_DIR = "/content/ComfyUI/models"

# Tạo thư mục đích nếu chưa có (ví dụ: models/unet, models/clip)
os.makedirs(f"{DEST_BASE_DIR}/unet", exist_ok=True)
os.makedirs(f"{DEST_BASE_DIR}/vae", exist_ok=True)
os.makedirs(f"{DEST_BASE_DIR}/clip", exist_ok=True)
os.makedirs(f"{DEST_BASE_DIR}/clip_vision", exist_ok=True)

# Danh sách các file và vị trí cần di chuyển (Nguồn -> Đích)
files_to_move = [
    ("unet", "wan2.1-i2v-14b-480p-Q4_0.gguf"),
    ("vae", "wan_2.1_vae.safetensors"),
    ("clip", "umt5_xxl_fp16.safetensors"),
    ("clip_vision", "clip_vision_h.safetensors"), # Đã sửa tên file clip vision
]

print("Bắt đầu di chuyển các file mô hình:")

for sub_dir, file_name in files_to_move:
    source_path = os.path.join(SOURCE_BASE_DIR, sub_dir, file_name)
    dest_path = os.path.join(DEST_BASE_DIR, sub_dir, file_name)

    if os.path.exists(source_path):
        # Sử dụng lệnh 'mv' để di chuyển (tương đương với cắt/dán)
        !mv "{source_path}" "{dest_path}"
        print(f"✅ Đã di chuyển: {file_name}")
    else:
        print(f"❌ Lỗi: Không tìm thấy file {file_name} tại {source_path}. Có thể đã tải thất bại hoặc đã được di chuyển trước đó.")

print("\n--- Hoàn tất di chuyển. Bạn có thể khởi động ComfyUI bây giờ. ---")
```
