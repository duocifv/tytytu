```python
# @title  ComfyUI view 1

%cd /content/ComfyUI

import threading, time, socket, torch, gc, psutil, os

# Giải phóng RAM & GPU bị chiếm
gc.collect()
torch.cuda.empty_cache()
os.system("fuser -k /dev/nvidia*")
print(f"🧠 RAM trống: {psutil.virtual_memory().available/1e9:.2f} GB")

# Mở iframe cho ComfyUI
def iframe_thread(port):
  while True:
      time.sleep(0.5)
      s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      if s.connect_ex(('127.0.0.1', port)) == 0:
          break
      s.close()
  from google.colab import output
  output.serve_kernel_port_as_iframe(port, height=1024)
  print("🌐 Mở trong cửa sổ mới tại:")
  output.serve_kernel_port_as_window(port)

threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

# Chạy ComfyUI tối ưu bộ nhớ
#!python main.py --listen --port 8188 --lowvram --async-offload --dont-upcast-attention --cpu-vae
!python main.py --listen --port 8188 --normalvram --disable-smart-memory --cache-none  --preview-method none --disable-metadata --async-offload

```

```python
# @title === CELL 1: SAO LƯU COMFYUI TOÀN BỘ VÀO GOOGLE DRIVE ===
from google.colab import drive
import os

# Mount Google Drive
drive.mount('/content/drive')

# Đường dẫn backup (bạn có thể đổi tên thư mục ở đây)
backup_dir = "/content/drive/MyDrive/Colab_Backups/ComfyUI_backup"

# Tạo thư mục nếu chưa có
os.makedirs(backup_dir, exist_ok=True)

# Sao lưu toàn bộ (trừ cache và output tạm)
!rsync -av --progress /content/ComfyUI/ "$backup_dir" --exclude 'output/' --exclude '__pycache__/' --exclude '*.tmp'

print("\n✅ Đã sao lưu xong toàn bộ ComfyUI vào Google Drive:")
print(backup_dir)
```
